```python
# Setup: separate X (308 CBFV features), y (both targets), and composition groups
import numpy as np
import pandas as pd
from sklearn.model_selection import GroupShuffleSplit

# X: 308 features after CBFV Oliynyk extraction
X = odf.drop(columns=['composition', 'formation_energy (eV/atom)', 'band_gap (eV)']).copy()
y_df = odf[['formation_energy (eV/atom)', 'band_gap (eV)']].copy()
comp_df = odf[['composition']].copy()

# Outer split: 80% train+val, 20% test (grouped by composition)
gss_outer = GroupShuffleSplit(n_splits=1, test_size=0.20, random_state=42)
trainval_idx, test_idx = next(gss_outer.split(X, y_df, groups=comp_df['composition']))

X_trainval = X.iloc[trainval_idx].reset_index(drop=True)
X_test     = X.iloc[test_idx].reset_index(drop=True)
y_trainval = y_df.iloc[trainval_idx].reset_index(drop=True)
y_test     = y_df.iloc[test_idx].reset_index(drop=True)
comp_trainval = comp_df.iloc[trainval_idx]['composition'].reset_index(drop=True)
comp_test     = comp_df.iloc[test_idx]['composition'].reset_index(drop=True)

# Inner split: from the 80% trainval, create train and val (e.g., 80/20 -> overall 64/16)
gss_inner = GroupShuffleSplit(n_splits=1, test_size=0.20, random_state=42)
train_idx, val_idx = next(gss_inner.split(X_trainval, y_trainval, groups=comp_trainval))

X_train = X_trainval.iloc[train_idx].reset_index(drop=True)
X_val   = X_trainval.iloc[val_idx].reset_index(drop=True)
y_train = y_trainval.iloc[train_idx].reset_index(drop=True)
y_val   = y_trainval.iloc[val_idx].reset_index(drop=True)
comp_train = comp_trainval.iloc[train_idx].reset_index(drop=True)
comp_val   = comp_trainval.iloc[val_idx].reset_index(drop=True)

# Select current target: band gap
ybg_train = y_train['band_gap (eV)']
ybg_val   = y_val['band_gap (eV)']
ybg_test  = y_test['band_gap (eV)']

# Optional sanity checks: no composition overlap across splits
assert set(comp_train).isdisjoint(set(comp_val))
assert set(comp_train).isdisjoint(set(comp_test))
assert set(comp_val).isdisjoint(set(comp_test))

# Shapes
print("Train shapes:", X_train.shape, ybg_train.shape)
print("Val shapes:",   X_val.shape,   ybg_val.shape)
print("Test shapes:",  X_test.shape,  ybg_test.shape)
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import mutual_info_regression

# Inputs assumed from previous step:
# X_train, X_val, X_test: feature DataFrames
# yfe_train, yfe_val, yfe_test: Series for formation energy

def compute_mi_ranking(X, y, n_neighbors=5, n_repeats=8, base_seed=42):
    """
    Compute mutual information scores for continuous target using multiple seeds, averaged.
    Fits only on provided X, y (should be training split).
    """
    feats = X.columns.to_list()
    all_runs = []
    for i in range(n_repeats):
        seed = base_seed + i
        mi = mutual_info_regression(
            X.values, y.values,
            n_neighbors=n_neighbors,
            random_state=seed
        )
        all_runs.append(mi)
    all_runs = np.vstack(all_runs)
    mi_mean = all_runs.mean(axis=0)
    mi_std  = all_runs.std(axis=0)
    out = pd.DataFrame({
        "feature": feats,
        "mi_mean": mi_mean,
        "mi_std": mi_std
    }).sort_values("mi_mean", ascending=False).reset_index(drop=True)
    out["rank"] = np.arange(1, len(out)+1)
    return out

mi_rank = compute_mi_ranking(X_train, ybg_train, n_neighbors=5, n_repeats=8, base_seed=42)

# Inspect top features
mi_rank.head(60)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,4))
plt.plot(mi_rank["rank"], mi_rank["mi_mean"], marker="o")
plt.xlabel("Feature Rank")
plt.ylabel("MI Score (mean)")
plt.title("Mutual Information vs Feature Rank")
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import pandas as pd
import numpy as np

scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train[final_selected_features])
X_val_scaled   = scaler_X.transform(X_val[final_selected_features])
X_test_scaled  = scaler_X.transform(X_test[final_selected_features])
print(X_train_scaled.shape)
print(X_val_scaled.shape)
print(X_test_scaled.shape)
```

```python
# Target scaling
ybg_train_array = ybg_train.values.reshape(-1,1)
ybg_val_array   = ybg_val.values.reshape(-1,1)
ybg_test_array  = ybg_test.values.reshape(-1,1)

scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(ybg_train_array).ravel()
y_val_scaled   = scaler_y.transform(ybg_val_array).ravel()
y_test_scaled  = scaler_y.transform(ybg_test_array).ravel()
```

```python
# -----------------------------
#  Define models
# -----------------------------
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor

models = {
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(alpha=1.0),
    'Lasso': Lasso(alpha=0.01),
    'RandomForest': RandomForestRegressor(n_estimators=200, random_state=42),
    'SVR': SVR(kernel='rbf', C=10, gamma='scale'),
    'AdaBoost': AdaBoostRegressor(n_estimators=200, random_state=42),
    'GradientBoosting': GradientBoostingRegressor(n_estimators=200, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=200, random_state=42, eval_metric='rmse', verbosity=0),
    'CatBoost': CatBoostRegressor(n_estimators=200, random_state=42, verbose=0),
    'LightGBM': LGBMRegressor(n_estimators=200, random_state=42)
}

# -----------------------------
# Train, predict, evaluate
# -----------------------------
results = []

for name, model in models.items():
    # Fit model
    model.fit(X_train_scaled, y_train_scaled)
    
    # Predict
    y_train_pred = scaler_y.inverse_transform(model.predict(X_train_scaled).reshape(-1,1)).ravel()
    y_val_pred   = scaler_y.inverse_transform(model.predict(X_val_scaled).reshape(-1,1)).ravel()
    y_test_pred  = scaler_y.inverse_transform(model.predict(X_test_scaled).reshape(-1,1)).ravel()
    
    # Metrics
    for split, y_true, y_pred in zip(['train','val','test'],
                                     [ybg_train, ybg_val, ybg_test],
                                     [y_train_pred, y_val_pred, y_test_pred]):
        r2   = r2_score(y_true, y_pred)
        mse  = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        mae  = mean_absolute_error(y_true, y_pred)
    

        results.append({
            'Model': name,
            'Split': split,
            'R2': r2,
            'MSE': mse,
            'MAE': mae,
            'RMSE': rmse
            
        })

# -----------------------------
#  Create results DataFrame
# -----------------------------
results_df = pd.DataFrame(results)
results_df
```

```python
# -----------------------------
# Imports
# -----------------------------
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd

# -----------------------------
# Define models
# -----------------------------
models = {
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'RandomForest': RandomForestRegressor(random_state=42),
    'SVR': SVR(kernel='rbf', C=10, gamma='scale'),
    'AdaBoost': AdaBoostRegressor(random_state=42),
    'GradientBoosting': GradientBoostingRegressor(random_state=42),
    'XGBoost': XGBRegressor(random_state=42, eval_metric='rmse', verbosity=0),
    'CatBoost': CatBoostRegressor(random_state=42, verbose=0),  # No n_estimators here
    'LightGBM': LGBMRegressor(random_state=42)
}

# -----------------------------
# Hyperparameter grids
# -----------------------------
param_grids = {
    'Ridge': {'alpha': [0.01, 0.1, 1, 10, 100]},
    'Lasso': {'alpha': [0.001, 0.01, 0.1, 1]},
    'RandomForest': {
        'n_estimators': [100, 200, 500],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10]
    },
    'AdaBoost': {
        'n_estimators': [100, 200, 500],
        'learning_rate': [0.01, 0.1, 0.5, 1.0]
    },
    'GradientBoosting': {
        'n_estimators': [100, 200, 500],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 10]
    },
    'XGBoost': {
        'n_estimators': [100, 200, 500],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 10]
    },
    'CatBoost': {
        'iterations': [100, 200, 500],  # use iterations instead of n_estimators
        'learning_rate': [0.01, 0.05, 0.1],
        'depth': [3, 5, 7]
    },
    'LightGBM': {
        'n_estimators': [100, 200, 500],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [-1, 5, 10]
    }
}

# -----------------------------
# Hyperparameter tuning loop
# -----------------------------
tuned_results = []

for name, model in models.items():
    print(f"Tuning {name}...")
    
    if name in param_grids:
        grid = GridSearchCV(model, param_grids[name], cv=3, scoring='r2', n_jobs=-1)
        grid.fit(X_train_scaled, y_train_scaled)
        best_model = grid.best_estimator_
        best_params = grid.best_params_
    else:
        # LinearRegression or models without grid
        best_model = model
        best_model.fit(X_train_scaled, y_train_scaled)
        best_params = "Default"
    
    # Predict on test set
    y_test_pred = scaler_y.inverse_transform(best_model.predict(X_test_scaled).reshape(-1,1)).ravel()
    
    # Metrics
    mse = mean_squared_error(ybg_test, y_test_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(ybg_test, y_test_pred)
    r2 = r2_score(ybg_test, y_test_pred)
    
    tuned_results.append({
        'Model': name,
        'R2': r2,
        'MSE': mse,
        'MAE': mae,
        'RMSE': rmse,
        'Best_Params': best_params
    })

# -----------------------------
# Results DataFrame
# -----------------------------
tuned_results_df = pd.DataFrame(tuned_results).sort_values(by='R2', ascending=False).reset_index(drop=True)
print(tuned_results_df)
```

```python
# -----------------------------
# 1. Export results to CSV
# -----------------------------
tuned_results_df.to_csv("magpie_model_tuning_bandgap_results.csv", index=False)
print("Tuning results saved to 'model_tuning_results.csv'")

# -----------------------------
# 2. Identify best model
# -----------------------------
best_model_info = tuned_results_df.iloc[0]  # Best R2
best_model_name = best_model_info['Model']
best_params = best_model_info['Best_Params']

print(f"Best Model: {best_model_name}")
print(f"Best Params: {best_params}")

# -----------------------------
# 3. Re-fit best model with best params
# -----------------------------
import joblib

# Dictionary mapping model names to classes
model_classes = {
    'LinearRegression': LinearRegression,
    'Ridge': Ridge,
    'Lasso': Lasso,
    'RandomForest': RandomForestRegressor,
    'SVR': SVR,
    'AdaBoost': AdaBoostRegressor,
    'GradientBoosting': GradientBoostingRegressor,
    'XGBoost': XGBRegressor,
    'CatBoost': CatBoostRegressor,
    'LightGBM': LGBMRegressor
}

# Handle default params case
if best_params == "Default":
    final_model = model_classes[best_model_name]()
else:
    if best_model_name == 'RandomForest' or best_model_name == 'AdaBoost' or best_model_name == 'GradientBoosting':
        final_model = model_classes[best_model_name](random_state=42, **best_params)
    elif best_model_name == 'XGBoost':
        final_model = XGBRegressor(random_state=42, eval_metric='rmse', verbosity=0, **best_params)
    elif best_model_name == 'CatBoost':
        final_model = CatBoostRegressor(random_state=42, verbose=0, **best_params)
    elif best_model_name == 'LightGBM':
        final_model = LGBMRegressor(random_state=42, **best_params)
    else:
        final_model = model_classes[best_model_name](**best_params)

# Fit on full training data
final_model.fit(X_train_scaled, y_train_scaled)

# -----------------------------
# 4. Save the model
# -----------------------------
model_filename = f"best_model_{best_model_name}.pkl"
joblib.dump(final_model, model_filename)
print(f"Best model saved as '{model_filename}'")
```
